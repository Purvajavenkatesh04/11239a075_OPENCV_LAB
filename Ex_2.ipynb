```python
print("Shape:", img.shape)
print("Size:", img.size)
print("Datatype:", img.dtype)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

plt.figure(figsize=(10,4))
plt.subplot(1,3,1); show(img, "Original")
plt.subplot(1,3,2); plt.imshow(gray, cmap='gray'); plt.title("Grayscale"); plt.axis('off')
plt.subplot(1,3,3); show(hsv, "HSV")
```

```python
b, g, r = cv2.split(img)
merged = cv2.merge([b, g, r])

plt.figure(figsize=(15, 5)) # Increased figsize to accommodate 4 subplots
plt.subplot(1,4,1); plt.imshow(b, cmap='gray'); plt.title("Blue"); plt.axis('off')
plt.subplot(1,4,2); plt.imshow(g, cmap='gray'); plt.title("Green"); plt.axis('off')
plt.subplot(1,4,3); plt.imshow(r, cmap='gray'); plt.title("Red"); plt.axis('off')
plt.subplot(1,4,4); show(merged, "Merged (BGR)") # Using 'show' for BGR to RGB conversion
plt.tight_layout()
plt.show()
```

```python
roi = img[100:300, 100:300]
resized = cv2.resize(img, (300, 300)) # Move this line from fKhdAkQOJ_Dm

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
show(roi, "ROI")
plt.subplot(1, 2, 2)
show(resized, "Resized Image") # Display resized image
plt.tight_layout()
plt.show()
```

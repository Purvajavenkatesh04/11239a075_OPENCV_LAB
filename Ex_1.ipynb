```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

def show(img, title="Image"):
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.axis('off')
```

**2. Read and Display Image**

```python
# Load sample image
img = cv2.imread('/EX 1.jpg')

show(img, "Original Image")
```

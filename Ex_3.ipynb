```python
# Need to reload img for Exercise 3 as the previous img might have been modified
img = cv2.imread('/img3.png') # Ensure original img for these operations

scaled = cv2.resize(img, None, fx=0.5, fy=0.5)

(h, w) = img.shape[:2]
center = (w//2, h//2)
matrix = cv2.getRotationMatrix2D(center, 45, 1.0)
rotated = cv2.warpAffine(img, matrix, (w, h))

crop = img[50:200, 50:200]

bright = cv2.convertScaleAbs(img, alpha=1.5, beta=50)

plt.figure(figsize=(16, 8)) # Adjust figure size as needed for 4 plots

plt.subplot(2, 2, 1)
show(scaled, "Scaled Image (fx=0.5, fy=0.5)")

plt.subplot(2, 2, 2)
show(rotated, "Rotated Image (45 deg)")

plt.subplot(2, 2, 3)
show(crop, "Cropped Image (50:200, 50:200)")

plt.subplot(2, 2, 4)
show(bright, "Brightness & Contrast (alpha=1.5, beta=50)")

plt.tight_layout()
plt.show()
```
